**3. From the Northeastern Seattle contact us webpage, web-scrape all email addresses. Use this webpage https://seattle.northeastern.edu/contact-us/ (Links to an external site.) (20 points)**

```python
# Received "HTTPError: HTTP Error 403: Forbidden." Determined that scraping was blocked by Cloud Flare
# from scraping the URL. Workaround from google search: 

# import packages
import re
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

# headers included to "mimic actual browser session" since NEU website is protected by Cloud Flare
req = Request("https://seattle.northeastern.edu/contact-us/", headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'})
neu_contact_us= str(urlopen(req).read().decode('utf-8'))

soup = BeautifulSoup(neu_contact_us)

print(soup)
```

```python
# Make the html code more readable
new_soup = BeautifulSoup((neu_contact_us), 'html.parser')
print(new_soup.prettify())
```

```python
# Find where the encoded (hidden) emails are located in the html: 
# The patter to find the hidden emails are that the emails 
# are the alphanumeric characters found after 'data-cfemail='

email_list = re.findall("data-cfemail\S+>", str(new_soup)) 
print(email_list)
```

```python
# We need to remove the extra characters
# i.e.: 'data-cfemail=,' and '">"'

email_list_remove_prefix = [y.replace('data-cfemail="','') for y in email_list]
print(email_list_remove_prefix)
```

```python
# Remove the unwanted '>' character in the email list
email_list_remove_bracket = [z.replace('">','') for z in email_list_remove_prefix]

print(email_list_remove_bracket)
```

```python
# Create new list of the encoded (hidden) emails
encoded_email_list = email_list_remove_bracket

print(encoded_email_list)
```

```python
# Since data on the NEU website is protected by a firewall: "Cloud Flare," 
# the data (email addresses) are obfuscated into string of alphanumeric digits (i.e.: encoded)

# Googled a function that will de-obfuscate the email address
# Source: https://stackoverflow.com/questions/59477061/email-protected-in-webscraping-how-to-get-the-text

def email(string):
    r = int(string[:2], 16)
    email = ''.join([chr(int(string[i:i+2], 16) ^ r)
                     for i in range(2, len(string), 2)])
    return email

print("The emails web scraped from the NEU Contact Us Page are: ")
# Loop through each encoded email and apply the function to de-code them
for idx, val in enumerate(encoded_email_list):
    print(email(val))

```
